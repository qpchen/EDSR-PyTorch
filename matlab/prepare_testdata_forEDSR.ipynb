After copy the LR images generated from using Prepare_TestData_HR_LR.m script to ../../dataset/benchmark/ dir,
use this to rename the image names to EDSR-PyTorch format.

Copy this file into the folder "LR_bicubic" under the dataset dir.

```python
import os
saveDir = './Prepare_EDSR_TestData'
LRDir = './LR'
HRDir = './HR'
HRList = os.listdir(HRDir)
LRList = os.listdir(LRDir)
```

```python
for dataset in HRList:
    if not os.path.exists(os.path.join(saveDir, dataset)):
        os.makedirs(os.path.join(saveDir,dataset), exist_ok=True)
```

```python
for degradation in LRList:
    degraDir = os.path.join(LRDir, degradation)
    for dataset in os.listdir(degraDir):
        dataDir = os.path.join(degraDir, dataset)
        scale_0 = os.listdir(dataDir)[0]
        imgHRDir = os.path.join(HRDir, dataset, scale_0)
        newImgHRDir = os.path.join(saveDir, dataset, 'HR')
        if not os.path.exists(newImgHRDir):
            os.makedirs(newImgHRDir, exist_ok=True)
        for img in os.listdir(imgHRDir):
            old = imgHRDir + os.sep + img
            new = newImgHRDir + os.sep + img.replace('_HR_{}'.format(scale_0), '')
            print('cp {} {}'.format(old, new))
            os.system('cp {} {}'.format(old, new))
        for scale in os.listdir(dataDir):
            imgLRDir = os.path.join(dataDir, scale)
            newImgLRDir = saveDir + os.sep + dataset + os.sep + 'LR_' + degradation.replace('LR', '') + os.sep + scale.replace('x', 'X')
            if not os.path.exists(newImgLRDir):
                os.makedirs(newImgLRDir, exist_ok=True)
            for img in os.listdir(imgLRDir):
                old = imgLRDir + os.sep + img
                new = newImgLRDir + os.sep + img.replace('_{}_'.format(degradation), '')
                print('cp {} {}'.format(old, new))
                os.system('cp {} {}'.format(old, new))
                
        
```
